این بالاترین سطح کانتینر برای تمام عناصر پلات است؛ در واقع، کل پنجره یا صفحه‌ای است که پلات در آن رسم می‌شود. یک Figure می‌تواند شامل یک یا چند "Axes" باشد و عناصر خاص دیگر مانند عنوان کلی شکل، راهنمای (legend) شکل و نوارهای رنگی (colorbar) را نیز مدیریت می‌کند.

     Axes (محورهای نمودار)
</div>

<div style="direction: rtl; text-align: right; font-size: 20px">
 این بخش که اغلب به آن "ساب‌پلات" یا "پلات" نیز گفته می‌شود، ناحیه واقعی‌ای است که داده‌ها در آن رسم می‌شوند. Axes به یک Figure متصل است و شامل عناصر ترسیم داده‌ها، از جمله محور X و محور Y (و محور Z برای پلات‌های سه‌بعدی) است. بیشتر متدهای ترسیم Matplotlib روی یک شیء Axes اعمال می‌شوند.

```python
fig, ax = plt.subplots() # ax = axes 
# You can make other plot with subplot like subplots(2)
ax.plot([1, 2, 3], [4, 5, 6])
```

Artist
</div>

<div style="direction: rtl; text-align: right; font-size: 20px">
    تعریف : به همه‌ی چیزهایی که روی Figure ظاهر می‌شن Artist می‌گن.
     این شامل خود اشیاء Figure، Axes و Axis، و همچنین عناصر خاصی مانند متن، خطوط، مجموعه‌ها (collections) و پچ‌ها (patches) می‌شود.

</div>

<div style="direction: ltr; text-align: left; font-size: 25px">
<pre>

<div style="direction: rtl; text-align: right; font-size: 25px">
    سبک MATLAB مانند (State-based / Pyplot style)
</div>

<div style="direction: rtl; text-align: right; font-size: 20px">
    این رابط شبیه به MATLAB است و برای ساخت سریع و آسان نمودارهای پایه مناسب است. بسیاری از توابع آن به صورت "حالت‌محور" (state-based) کار می‌کنند، به این معنی که تغییرات روی "نمودار فعلی" اعمال می‌شوند.
    <br>
    در این روش، Figure و Axes به‌صورت خودکار ساخته می‌شن.
</div>

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("simple")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<div style="direction: rtl; text-align: right; font-size: 25px">
    سبک شی‌گرا (Object-oriented style)
</div>

<div style="direction: rtl; text-align: right; font-size: 20px">
    این رابط قدرتمندتر و انعطاف‌پذیرتر است و به شما کنترل کامل‌تری بر روی تمام اجزای نمودار می‌دهد. در این رویکرد، شما به طور صریح اشیاء Figure (کل فضای نمودار) و Axes (محدوده رسم واقعی نمودار) را ایجاد کرده و سپس متدها را روی این اشیاء فراخوانی می‌کنید. این روش برای نمودارهای پیچیده‌تر و سفارشی‌سازی‌های عمیق‌تر توصیه می‌شود.
    <br>
    این روش حرفه‌ای‌تره. خودت به‌صورت دقیق Figure و Axes رو می‌سازی و روی اون‌ها کار می‌کنی.
</div>

```python
fig, ax = plt.subplots()  # ساخت Figure و Axes
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("oo")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

<div style="direction: rtl; text-align: right; font-size: 30px">
    ساخت تابع کمکی (Helper Function)
</div>

<div style="direction: rtl; text-align: right; font-size: 20px">
وقتی نمودارهای مشابه زیادی می‌خوای رسم کنی، بهتره یک تابع کمکی تعریف کنی که کار تکراری رو انجام بده. این‌طوری کدت:
    <br>
⦁    تمیزتر می‌شه
    <br>
⦁    قابل‌تکرار و قابل‌استفاده در پروژه‌های بزرگ‌تر می‌شه
</div>

```python
def my_plot(ax, data, title):
    ax.plot(data)
    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
```

<div style="direction: rtl; text-align: right; font-size: 20px">
     استفاده از این تابع
</div>

```python
fig, axs = plt.subplots(2)  # دو محور (Axes) می‌سازه

my_plot(axs[0], [1, 2, 3], "first")
my_plot(axs[1], [3, 2, 1], "second")

plt.tight_layout()
plt.show()
```

<div style="direction: rtl; text-align: right; font-size: 30px">
     نوشتن فرمول‌های ریاضی در متن (LaTeX-style)
</div>

<div style="direction: rtl; text-align: right; font-size: 20px">
    Matplotlib از یک زیرمجموعهٔ LaTeX پشتیبانی می‌کنه. برای نوشتن فرمول، فقط کافیه اون رو داخل r"$ ... $" بنویسی.
    <br>
    r قبل از رشته یعنی "رشته خام" (raw string)، تا بک‌اسلش‌ها خراب نشن.
</div>

```python
plt.title(r"$E = mc^2$")
plt.xlabel(r"$x$")
plt.ylabel(r"$\alpha + \beta$")
```

    <br>

    <br>
